```python
from telethon import TelegramClient
import logging
import asyncio
import nest_asyncio
import os
import csv
```

```python
# Apply nest_asyncio to allow nested event loops
nest_asyncio.apply()

# Setup logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# Replace these with your own values
api_id = '21747532' 
api_hash = '96e9bac67654b8b698658d191a3fad1e'  

client = TelegramClient('tel_session_name', api_id, api_hash)

# Create data directory if it doesn't exist
data_dir = 'data'
if not os.path.exists(data_dir):
    os.makedirs(data_dir)

# CSV file path
csv_file_path = os.path.join(data_dir, 'data.csv')

# Initialize CSV file with headers
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Date', 'Sender ID', 'Message'])

async def scrape_channel(channel_username):
    try:
        await client.start()
        channel = await client.get_entity(channel_username)
        async for message in client.iter_messages(channel):
            # Log message details
            logger.info(f"Scraping message from {channel_username}")

            # Save message to CSV
            with open(csv_file_path, 'a', newline='', encoding='utf-8') as csvfile:
                csv_writer = csv.writer(csvfile)
                csv_writer.writerow([message.date, message.sender_id, message.text])

            # Download images if present
            if message.photo:
                image_path = os.path.join(data_dir, 'images')
                if not os.path.exists(image_path):
                    os.makedirs(image_path)
                await message.download_media(file=image_path)
        logger.info(f'Successfully scraped {channel_username}')
    except Exception as e:
        logger.error(f"Error scraping {channel_username}: {e}")

async def main():
    channels = [
        'yetenaweg',
        'CheMed123',
        'lobelia4cosmetics',
    ]
    for channel in channels:
        await scrape_channel(channel)

# Run the event loop
if __name__ == '__main__':
    asyncio.run(main())
```
